Q.4.What is Prompt Engineering, and explain any two types of prompts used in Generative AI?

Ans:- Prompt Engineering means writing good instructions, so that an AI model gives the best and most accurate output.It is like asking a very smart assistant a clear question so you get a clear answer.Better prompt gives Better AI response.

The two types of promts are as follows:-

1.Instruction Prompt:- In this type of promt we tell the AI exactly what to do.

for example:-Write a short caption for Instagram.

2.Example-Based Prompt (Few-shot Prompt):- In this type of promt we give examples to show the AI what kind of answer we want.

for example:-

Good morning → Suprabhat

Thank you → Dhanyavaad

Now translate: How are you.

Q.5.What are the key differences between a Business Analyst and a Data Analyst in terms of roles, responsibilities, and focus?

Ans:- The difference between Business Analyst and Data Analyst are as follows:-

-> Business Analyst.

1.Focuses on business problems, processes, goals, and solutions.

2.Talks to clients/stakeholders

3.Understands business needs

4.Creates requirements and business plans

5.Suggests improvements for business processes


Business Analyst usees skills like:-Communication,Documentation,Understanding business processes,Problem-solving etc.


-> Data Analyst.

1.Focuses on data, numbers, patterns, and insights.

2.Collects data

3.Cleans and analyzes data

4.Creates charts, dashboards, and reports

5.Finds trends or patterns in numbers

Data Analyst works on:-Excel,SQL,Data cleaning,Statistics,Visualization tools (Power BI, Tableau).

Q.8.Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.

Ans:- The stages from descriptive to generative are as follows:-

1.Descriptive Analytics What happened?:-This is the starting stage.It tells us past information using reports and charts.

2.Diagnostic Analytics Why did it happen?:-It finds the reason behind the results.

3.Predictive Analytics What will happen?:-Uses data + patterns + machine learning to predict the future.

4.Prescriptive Analytics What should we do?:-Gives suggestions or actions to improve results.

5.Generative AI in this  AI create solutions:-AI not only analyzes but also creates content, ideas, and solutions on its own.
